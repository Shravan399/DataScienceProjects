## Data Preperation

#### step-1:Identify the inputs(X) and output(y)

```python
X = df[['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z']]
y = df['price']  
```

#### step-2: split the data into train and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=100) 
```

```python
X_train.head()
```

```python
# checking shape of train and test data
print(X_train.shape, y_train.shape)

print(X_test.shape, y_test.shape) 
```

## separating categorical and Numerical columns:

```python
X_train.head() 
```

```python
X_train.dtypes 
```

```python
# categorical coulmns
X_train_cat = X_train.select_dtypes(include=['object'])

X_train_cat.head() 
```

```python
# Numerical columns
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])

X_train_num.head() 
```

## Scaling the numerical columns

```python
X_train_num.head() 
```

```python
# scaling the numerical features using standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head() 
```

```python
# checking mean,median,standard deviation of numerical columns after rescaling
X_train_num.describe() 
```

```python
print("Number of Numerical Features:", scaler.n_features_in_)
print("Mean of each column:", scaler.mean_)
print("Std of each column:", np.sqrt(scaler.var_))  
```

## Applying Label encoding on categorical columns:

```python
# checking value_counts for 'cut' column using normaization
X_train_cat['cut'].value_counts(normalize=True) 
```

```python
# checking value_counts for 'color' column using normalization
X_train_cat['color'].value_counts(normalize=True) 
```

```python
# checking value_counts for 'clarity' column using normalization
X_train_cat['clarity'].value_counts(normalize=True)
```

```python
# getting index of cut column
X_train_cat_le = pd.DataFrame(index=X_train_cat.index)

X_train_cat_le.head() 
```

```python
# getting unique values in the cut columns
X_train_cat.cut.unique() 
```

```python
# Applying label encoding for cut column
cut_encoder = {'Fair' : 1, 'Good' : 2, 'Very Good' : 3, 'Ideal' : 4, 'Premium' : 5}

X_train_cat_le['cut'] = X_train_cat['cut'].apply(lambda x : cut_encoder[x])

X_train_cat_le.head() 
```

```python
# Getting unique values In the color column
X_train_cat.color.unique() 
```

```python
# Applying label encoding for color column
color_encoder = {'J':1, 'I':2, 'H':3, 'G':4, 'F':5, 'E':6, 'D':7}

X_train_cat_le['color'] = X_train_cat['color'].apply(lambda x : color_encoder[x])

X_train_cat_le.head() 
```

```python
# Getting unique values in the clarity column
X_train_cat.clarity.unique() 
```

```python
# Applying label encoding for clarity column
clarity_encoder = {'I1':1, 'SI2':2, 'SI1':3, 'VS2':4, 'VS1':5, 'VVS2':6, 'VVS1':7, 'IF':8}

X_train_cat_le['clarity'] = X_train_cat['clarity'].apply(lambda x : clarity_encoder[x])

X_train_cat_le.head() 
```

## Concatinating the Encoded Categorical Features and Rescaled Numerical Features:

```python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_le], axis=1)

X_train_transformed.head()
```

## Preparing test data

```python
X_test.head() 
```

```python
# checking info of the X_test data
X_test.info() 
```

```python
# separating categorical columns
X_test_cat = X_test.select_dtypes(include=['object'])

X_test_cat.head() 
```

```python
# separarting numerical columns of test data
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])

X_test_num.head() 
```

```python
# rescaling numerical columns of test data
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = X_test_num.columns, 
                                   index = X_test_num.index)

X_test_num_rescaled.head()  
```

```python
X_test_cat_le = pd.DataFrame(index = X_test_cat.index)

X_test_cat_le.head() 
```

```python
# applying label encoding for categorical columns of test data
X_test_cat_le['cut'] = X_test_cat['cut'].apply(lambda x : cut_encoder[x])

X_test_cat_le['color'] = X_test_cat['color'].apply(lambda x : color_encoder[x])

X_test_cat_le['clarity'] = X_test_cat['clarity'].apply(lambda x : clarity_encoder[x])

X_test_cat_le.head() 
```

```python
# concatinating categorical and numerical columns of test data
X_test_transformed = pd.concat([X_test_num_rescaled, X_test_cat_le], axis=1)

X_test_transformed.head() 
```

## Building a model using KNN Regression

```python
x_train = X_train_transformed.to_numpy()
x_test = X_test_transformed.to_numpy()
np_y_train = y_train.to_numpy()

def Algorithm(x_train, np_y_train, k, xq):
    distances = np.sqrt(np.sum((x_train - xq)**2, axis=1))
    sorted_distances_indices = np.argsort(distances)[:k]
    return np.mean(np_y_train[sorted_distances_indices])

k = 5  
xq = x_test
df_list = []

for i in range(len(xq)):
    prediction = Algorithm(x_train, np_y_train, k, xq[i])
    df_list.append(prediction)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': df_list})

temp_df.head() 
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, df_list))

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, df_list))

print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, df_list))) 
```
